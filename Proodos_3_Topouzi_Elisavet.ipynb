# Δημιουργία βασικού χάρτη

Δώστε συντεταγμένες (latitude and longitude) για την περιοχή της Ελλάδας. Μπορείτε ακολουθώντας τις παρακάτω οδηγίες να αναζητήσετε μια πόλη πχ Αθήνα και να χρησιμοποιήσετε το ανάλογο zoom για να καλύψετε την ζητούμενη περιοχή. 

**Example**

For example, a quick Google search tells us that the **coordinates of Fora's venue** are `51.524538` `-0.100565`. We decide that the **map tiles** should be displayed with `OpenStreetMap` layout, with an **initial zoom** value of `17`.

```python
lat = 40.00000
lon = 28.00000
my_zoom = 6
my_tiles = "OpenStreetMap"

my_map = folium.Map(location=[lat, lon], tiles=my_tiles, zoom_start=my_zoom)
```

# Χάρτης Φύσης και Βιοποικιλότητας

Στην διεύθυνση http://ekbygis.biodiversity-info.gr/map/index.html υπάρχει ο Χάρτης Φύσης και Βιοποικιλότητας όπου μπορείτε να εντοπίσετε τους 10 Ελληνικούς υγρότοπους της Συνθήκης Ραμσάρ. Αφού τους επιλέξετε από το αριστερό μενού, βρείτε τις συντεταγμένες τους και ακολουθόντας τις παρακάτω οδηγίες δημιουργήστε markers σε κάθε μια θέση. 

**Example**

Let's say that we want to highlight Fora's location on the map. We can do this by **adding a Marker** to the existing map. We also want to make clear that we're pointing to Decoded's offices, so we can **use a customised icon**.

```python
# Παράδειγμα 'Ορεστιάδα'
my_icon = folium.Icon(icon='ok', color='green', icon_color='lightgreen', prefix='glyphicon')

# use the add_to method to add the Marker object to the map
lat_orest = 41.506950
lon_orest = 26.531530
folium.Marker([lat_orest, lon_orest], icon=my_icon).add_to(my_map)

my_map
```

# Προσθέστε πληροφορίες pop-up

**Example**

To make sure nobody misses out on the social event, we **add a second marker** indicating the location of the drinks venue.

To remind everyone about the scheduled time, we **create an interactive pop-up** displaying some additional information.

```python
#Δέλτα Έβρου
lon_evrou = 26.03850
lat_evrou = 40.79510

ev_icon = folium.Icon(icon='ok', color='cadetblue', icon_color='lightblue', prefix='glyphicon')
folium.Marker([lat_evrou, lon_evrou],
          	icon=ev_icon,
          	tooltip='Πληροφορίες!',
          	popup='<b>Δέλτα Έβρου!</b> </br> 9,267 εκτάρια.'
         	).add_to(my_map)

# Λίμνη Βιστονίδα
lat_bist = 40.96260
lon_bist = 25.28730

bist_icon = folium.Icon(icon='ok', color='cadetblue', icon_color='lightblue', prefix='glyphicon')
folium.Marker([lat_bist, lon_bist],
          	icon=bist_icon,
          	tooltip='Πληροφορίες!',
          	popup='<b>Λίμνη Βιστονίδα,Πόρτο-Λάγος!</b> </br> 24,396 εκτάρια.'
         	).add_to(my_map)

# Δέλτα Νέστου
lat_nestou = 40.91804
lon_nestou = 24.76271

nest_icon = folium.Icon(icon='ok', color='cadetblue', icon_color='lightblue', prefix='glyphicon')
folium.Marker([lat_nestou, lon_nestou],
          	icon=nest_icon,
          	tooltip='Πληροφορίες!',
          	popup='<b>Δέλτα Νέστου!</b> </br> 21,930 εκτάρια.'
         	).add_to(my_map)
          
# Λίμνη Βόλβη και Κορώνεια
lat_bol = 40.67989
lon_bol = 23.37003

bol_icon = folium.Icon(icon='ok', color='cadetblue', icon_color='lightblue', prefix='glyphicon')
folium.Marker([lat_bol, lon_bol],
          	icon=bol_icon,
          	tooltip='Πληροφορίες!',
          	popup='<b>Λίμνη Βόλβη και Κορώνεια!</b> </br> 16,388 εκτάρια.'
         	).add_to(my_map)
          
# Τεχνητή Λίμνη Κερκίνη
lat_kerk = 41.20969
lon_kerk = 23.14604

kerk_icon = folium.Icon(icon='ok', color='cadetblue', icon_color='lightblue', prefix='glyphicon')
folium.Marker([lat_kerk, lon_kerk],
          	icon=kerk_icon,
          	tooltip='Πληροφορίες!',
          	popup='<b>Τεχνητή Λίμνη Κερκίνης!</b> </br> 10,966 εκτάρια.'
         	).add_to(my_map)

# Δέλτα Αξιού,Λουδία,Αλιάκμονα
lat_ax = 40.47591
lon_ax = 22.61886

ax_icon = folium.Icon(icon='ok', color='cadetblue', icon_color='lightblue', prefix='glyphicon')
folium.Marker([lat_ax, lon_ax],
          	icon=ax_icon,
          	tooltip='Πληροφορίες!',
          	popup='<b> Δέλτα Αξιού,Λουδία,Αλιάκμονα!</b> </br> 11,808 εκτάρια.'
         	).add_to(my_map)

# Λίμνη Μικρή Πρέσπα
lat_pres = 40.77740
lon_pres = 21.10255

pres_icon = folium.Icon(icon='ok', color='cadetblue', icon_color='lightblue', prefix='glyphicon')
folium.Marker([lat_pres, lon_pres],
          	icon=pres_icon,
          	tooltip='Πληροφορίες!',
          	popup='<b> Λίμνη Μικρή Πρέσπα!</b> </br> 5,078 εκτάρια.'
         	).add_to(my_map)    
                  
# Αμβρακικός Κόλπος
lat_am  = 39.04045
lon_am =  20.93606

am_icon = folium.Icon(icon='ok', color='cadetblue', icon_color='lightblue', prefix='glyphicon')
folium.Marker([lat_am, lon_am],
          	icon=am_icon,
          	tooltip='Πληροφορίες!',
          	popup='<b> # Αμβρακικός Κόλπος!</b> </br> 65,400 εκτάρια.'
         	).add_to(my_map)    

 # Λιμνοθάλασσα Μεσολογγίου
lat_mes = 38.34650
lon_mes = 21.32047

mes_icon = folium.Icon(icon='ok', color='cadetblue', icon_color='lightblue', prefix='glyphicon')
folium.Marker([lat_mes, lon_mes],
          	icon=mes_icon,
          	tooltip='Πληροφορίες!',
          	popup='<b> # Λιμνοθάλασσα Μεσολογγίου!</b> </br> 2,200 εκτάρια.'
         	).add_to(my_map)
          
# Λιμνοθάλασσα Κοτυχίου
lat_kot = 38.07964
lon_kot = 21.34649

kot_icon = folium.Icon(icon='ok', color='cadetblue', icon_color='lightblue', prefix='glyphicon')
folium.Marker([lat_kot, lon_kot],
          	icon=kot_icon,
          	tooltip='Πληροφορίες!',
          	popup='<b> # Λιμνοθάλασσα Κοτυχίου!</b> </br> 6,302 εκτάρια.'
         	).add_to(my_map)

my_map
```
